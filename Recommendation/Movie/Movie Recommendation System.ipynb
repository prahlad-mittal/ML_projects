```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
column_names = ["item_id", "movie_title", "release date", "video release date", "IMDb URL", "unknown", "Action", "Adventure", "Animation", "Children's", "Comedy", "Crime", "Documentary", "Drama", "Fantasy", "Film-Noir", "Horror", "Musical", "Mystery", "Romance", "Sci-Fi", "Thriller", "War", "Western"]
item = pd.read_csv("ml-100k/u.item", sep='|', names = column_names, encoding='latin-1')
item
```

```python
print(item.info())
```

```python
index_movie_dict = {}
for i in range(len(item)):
    index_movie_dict[i] = item.loc[i]['movie_title']
```

```python
index_movie_dict
```

```python
movie_index_dict = {}
for i in range(len(item)):
    movie_index_dict[item.loc[i]['movie_title']] = i
```

```python
movie_index_dict
```

```python
item.drop(["item_id", "movie_title", "release date", "video release date", "IMDb URL"], axis = 1, inplace = True)
item
```

```python
item.columns
```

```python
genres = item.to_numpy()
```

```python
genres
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel
genres_similarity_matrix = sigmoid_kernel(genres, genres)
print(genres_similarity_matrix)
```

```python
genres_similarity_matrix.shape
```

```python
print(genres_similarity_matrix[0])
```

```python
index_movie_dict.values()
```

```python
def getRecommendations(movie_name):
    idx = movie_index_dict[movie_name]
    similarity_list = genres_similarity_matrix[idx]
    lst = []
    for i in range(len(similarity_list)):
        lst.append((similarity_list[i],i))
    return sorted(lst, reverse = True)
```

# Get recommendations corresponding to a movie

```python
recommendations_list = getRecommendations('Star Wars (1977)')
for element in recommendations_list:
    print(index_movie_dict[element[1]])
```

# Resolving the issue of having the recently watched movie in the recommendation list

```python
recommendations_list = getRecommendations('Star Wars (1977)')
final_recommendation_list = []
for element in recommendations_list:
    final_recommendation_list.append(index_movie_dict[element[1]])
final_recommendation_list.remove('Star Wars (1977)')
final_recommendation_list[:10]
```

```python
def recommend_movies(movie_title, number_of_recommendations):
    recommendations_list = getRecommendations(movie_title)
    final_recommendation_list = []
    for element in recommendations_list:
        final_recommendation_list.append(index_movie_dict[element[1]])
    final_recommendation_list.remove(movie_title)
    for movie in final_recommendation_list[:number_of_recommendations]:
        print(movie)
```

```python
recommend_movies('Star Wars (1977)', 10)
```

```python
recommend_movies('Lion King, The (1994)', 10)
```

```python
recommend_movies('Jurassic Park (1993)', 20)
```
