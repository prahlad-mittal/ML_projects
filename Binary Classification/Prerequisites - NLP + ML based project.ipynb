```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import sigmoid_kernel

from pywebio.input import *
from pywebio.output import *
from pywebio.platform.flask import webio_view
from pywebio import STATIC_PATH
from flask import Flask, send_from_directory
import argparse
from pywebio import start_server
app = Flask(__name__)

column_names = ["item_id", "movie_title", "release date", "video release date", "IMDb URL", "unknown", 
                "Action", "Adventure", "Animation", "Children's", "Comedy", "Crime", "Documentary", 
                "Drama", "Fantasy", "Film-Noir", "Horror", "Musical", "Mystery", "Romance", "Sci-Fi", 
                "Thriller", "War", "Western"]
item = pd.read_csv("u.item", sep='|', names = column_names, encoding='latin-1')
genres = pd.read_csv("u.item", sep='|', names = column_names, encoding='latin-1')
genres.drop(["item_id", "movie_title", "release date", "video release date", "IMDb URL"], axis = 1, inplace = True) 
matrix = genres.to_numpy()
similarity_matrix = sigmoid_kernel(matrix, matrix)
indices = pd.Series(item.index, index = item['movie_title']).drop_duplicates()

def get_recommendations(title, similarity_matrix = similarity_matrix):
    idx = indices[title]
    similarity_scores_for_a_particular_movie = list(enumerate(similarity_matrix[idx]))
    similarity_scores_for_a_particular_movie = sorted(similarity_scores_for_a_particular_movie, key = lambda x:x[1], reverse = True)
    similarity_scores_for_a_particular_movie = similarity_scores_for_a_particular_movie[1:11]
    recommended_movie_indices = [i[0] for i in similarity_scores_for_a_particular_movie]
    return item['movie_title'].iloc[recommended_movie_indices]

def recommend():
    img = open('movies.jpg', 'rb').read()
    put_image(img, width='900px')
    put_markdown('# **Movie Recommendation System by `Rishabh Bafna`**')
    put_html('<h1 style="font-size:20px; text-align:center; color:blue; background-color:powderblue;">Hello World! Welcome!<br></h1>')
    recommendation_system = select('Which type of recommendation system would you prefer?', ['Generalized Version (Preferred)', 'Collaborative Filtering', 'Content-based Filtering', 'Simple Recommendation System', 'Deep Learning-based Recommendation System'])
    if(recommendation_system == 'Collaborative Filtering'):
        recommendation_system = select('Which type of Collaborative Filtering approach would you prefer to get recommendations?', ['Item-to-Item Collaborative Filtering with the usage of Cosine Similarity', 'Item-to-Item Collaborative Filtering with the usage of Pearson Correlation', 'Item-to-Item Collaborative Filtering with the usage of Kendall rank correlation', 'Item-to-Item Collaborative Filtering with the usage of Spearman correlation'])
    if(recommendation_system == 'Content-based Filtering'):
        recommendation_system = select('Which type of Content-based Filtering approach would you prefer to get recommendations?', ['Content-based Filtering with the usage of sigmoid_kernel', 'Content-based Filtering with the usage of Cosine Similarity'])
    
    #Create separate function for the following Algorithm 5: Content-based Filtering with the usage of sigmoid_kernel
    entered_movie = input("Enter one of the movies from the database", type=TEXT)
    recommendations = get_recommendations(entered_movie)
    final_recommendations_list = recommendations.tolist()
    if(entered_movie in final_recommendations_list):
        final_recommendations_list.remove(entered_movie)
    put_html('<hr>')
    put_markdown("Recommendations for you similar to `%s` movie are as follows: " % entered_movie)
    put_html('<hr>')
    for movie in final_recommendations_list:
        put_text("%s" % movie)

#app.add_url_rule('/rishabh', 'webio_view', webio_view(recommend), methods=['GET', 'POST', 'OPTIONS'])
#app.run(host='localhost', port=80)
app.add_url_rule('/rishabh', 'webio_view', webio_view(recommend), methods=['GET', 'POST', 'OPTIONS'])

if __name__ == '__main__':
    parser = argparse.ArgumentParser()
    parser.add_argument("-p", "--port", type=int, default=8080)
    args = parser.parse_args()
    start_server(recommend, port=args.port)
#if __name__ == '__main__':
    #recommend()

#app.run(host='localhost', port=80)

#visit http://localhost/rishabh to open the PyWebIO application.
```
