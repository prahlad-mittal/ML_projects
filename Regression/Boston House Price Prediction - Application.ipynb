```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import sigmoid_kernel
from pywebio.input import *
from pywebio.output import *
import time
from pywebio.platform.flask import webio_view
from pywebio import STATIC_PATH
from flask import Flask, send_from_directory
app = Flask(__name__)

from sklearn.datasets import load_boston
boston = load_boston()

df = pd.DataFrame(boston['data'], columns = boston['feature_names'])
df['target'] = boston['target']

X = boston['data']
y = boston['target']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 100)

from xgboost import XGBRegressor
xgboostRegressor = XGBRegressor()
xgboostRegressor.fit(X, y)

def predict():
    user_input = []
    crim = input("Enter per capita crime rate by town (CRIM) ", type = FLOAT)
    user_input.append(crim)
    zn = input("Enter proportion of residential land zoned for lots over 25,000 sq.ft.", type = FLOAT)
    user_input.append(zn)
    indus = input("Enter proportion of non-retail business acres per town ", type = FLOAT)
    user_input.append(indus)
    chas = input("Enter Charles River dummy variable (= 1 if tract bounds river; 0 otherwise ", type = FLOAT)
    user_input.append(chas)
    nox = input("Enter nitric oxides concentration (parts per 10 million) ", type = FLOAT)
    user_input.append(nox)
    rm = input("Enter average number of rooms per dwelling ", type = FLOAT)
    user_input.append(rm)
    age = input("Enter proportion of owner-occupied units built prior to 1940 ", type = FLOAT)
    user_input.append(age)
    dis = input("Enter weighted distances to five Boston employment centres ", type = FLOAT)
    user_input.append(dis)
    rad = input("Enter index of accessibility to radial highways ", type = FLOAT)
    user_input.append(rad)
    tax = input("Enter full-value property-tax rate per $10,000 ", type = FLOAT)
    user_input.append(tax)
    ptratio = input("Enter pupil-teacher ratio by town ", type = FLOAT)
    user_input.append(ptratio)
    b = input("Enter the value for 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town ", type = FLOAT)
    user_input.append(b)
    lstat = input("Enter % lower status of the population ", type = FLOAT)
    user_input.append(lstat)
    prediction = xgboostRegressor.predict([user_input])
    
    put_html('<hr>')
    put_markdown("Predicted house price is as follows: `%s` " % str(prediction))
    put_html('<hr>')

app.add_url_rule('/houseprice', 'webio_view', webio_view(predict), methods=['GET', 'POST', 'OPTIONS'])
app.run(host='localhost', port=80)
```
